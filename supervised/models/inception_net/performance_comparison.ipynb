# Performance Comparison

```python
import os
import json
import matplotlib.pyplot as plt
```

```python
histories = [f for f in os.listdir('.') if '.json' in f]
```

```python
histories
```

```python
with open(histories[-2], 'r') as h:
    print(json.load(h)['val_accuracy'])
```

```python
acc_dist = {'acc': [], 'label': []}
for hist in histories:
    with open(hist, 'r') as h:
        print(hist)
        acc = json.load(h)['val_accuracy']
        acc_dist['acc'].append(acc)
        acc_dist['label'].append(hist)
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
for acc, label in zip(acc_dist['acc'], acc_dist['label']):
    x = [i+1 for i in range(len(acc))]
    plt.plot(x, acc, label=label)
plt.legend()
plt.show()
```
